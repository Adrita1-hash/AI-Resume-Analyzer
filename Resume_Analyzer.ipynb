```python
import pdfplumber

resume_path = "resumes/RESUME.pdf"

text = ""

with pdfplumber.open(resume_path) as pdf:
    for page in pdf.pages:
        extracted = page.extract_text()

        if extracted:
            text += extracted

print(text)
```

```python
cleaned_text = text.lower()

print(cleaned_text)
```

```python
import re

cleaned_text = re.sub(r'[^a-zA-Z\s]', '', cleaned_text)

print(cleaned_text)
```

```python
import nltk

nltk.download('punkt_tab')
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

nltk.download('punkt')
nltk.download('stopwords')

stop_words = set(stopwords.words('english'))

words = word_tokenize(cleaned_text)

filtered_words = []

for word in words:
    if word not in stop_words:
        filtered_words.append(word)

print(filtered_words)
```

```python
final_text = " ".join(filtered_words)

print(final_text)
```

```python
skills = [
    "python",
    "java",
    "sql",
    "mysql",
    "postgresql",
    "html",
    "css",
    "javascript",
    "react",
    "machine learning",
    "data structures",
    "algorithms",
    "nlp",
    "pandas",
    "scikit-learn"
]

print(skills)
```

```python
detected_skills = []

for skill in skills:
    if skill in final_text:
        detected_skills.append(skill)

print("Detected Skills:")
print(detected_skills)
```

```python
job_description = """
We are looking for a Python Developer with knowledge of SQL,
React, Machine Learning, HTML, CSS, and JavaScript.
Candidates should have strong problem-solving skills
and understanding of data structures and algorithms.
"""

print(job_description)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
documents = [final_text, job_description.lower()]

tfidf = TfidfVectorizer()

matrix = tfidf.fit_transform(documents)

similarity_score = cosine_similarity(matrix[0:1], matrix[1:2])

score = similarity_score[0][0] * 100

print("ATS Match Score:")
print(round(score, 2), "%")
```

```python
job_skills = []

job_description_lower = job_description.lower()

for skill in skills:
    if skill in job_description_lower:
        job_skills.append(skill)

print("Job Required Skills:")
print(job_skills)
```

```python
matched_skills = []

for skill in detected_skills:
    if skill in job_skills:
        matched_skills.append(skill)

print("Matched Skills:")
print(matched_skills)

match_percentage = (len(matched_skills) / len(job_skills)) * 100

print("\nSkill Match Percentage:")
print(round(match_percentage, 2), "%")
```

```python
job_skills = []

job_description_lower = job_description.lower()

for skill in skills:
    if skill in job_description_lower:
        job_skills.append(skill)

print("Job Required Skills:")
print(job_skills)
```

```python
missing_skills = []

for skill in job_skills:
    if skill not in detected_skills:
        missing_skills.append(skill)

print("Missing Skills:")
print(missing_skills)
```

```python
print("========== ATS RESUME ANALYSIS REPORT ==========\n")

print("Detected Skills:")
print(detected_skills)

print("\nJob Required Skills:")
print(job_skills)

print("\nMatched Skills:")
print(matched_skills)

print("\nMissing Skills:")
print(missing_skills)

print("\nSkill Match Percentage:")
print(round(match_percentage, 2), "%")
```
